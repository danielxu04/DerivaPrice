## Markowitz Model Demonstration

```python
from MarkowitzModel import *
```

```python
stocks = ['GOOG', 'MSFT', 'AAPL', 'AMZN', 'NVDA', 'AMD', 'TSLA']    
start_date = '2013-09-01'
end_date = '2023-09-01'
```

### Download Data

Download stock data from 2013-09-01 to 2023-09-01 and then plot the results. 

```python
data = download_data(start_date, end_date, stocks)
plot_data(data)
```

### Statistics for Returns

Below we will compute (very) basic statistics for the log returns of the data: annual mean and the covariance matrix of returns.

```python
returns = compute_returns(data)
compute_statistics(returns)
```

### Generate Random Portfolios

```python
weights, means, risks = generate_portfolios(returns, stocks)
plot_portfolios(means, risks)
```

This scatterplot correlates Expected Volatility and Expected Returns, with a Sharpe Ratio colour bar. Our graph demonstrates that the expected return is positively correlated with expected volatility. In other words, as the expected volatility increases, the expected return tends to increase as well.

Also, observe that a steeper slope correlates with a higher Sharpe Ratio.

### Portfolio Optimization

```python
optimum = optimize_portfolio(weights, returns, stocks)
plot_optimal_portfolio(optimum, returns, means, risks)
```

In conclusion, our optimal portfolio is highlighted with the 'green star marker', evinced by the Markowitz model for portfolio optimization.
